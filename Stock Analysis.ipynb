```python
import pandas as pd

import AlphaVantage
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Key = 'Z0DPEH5EFNHR6W8D'
IBMId = 'IBM'
AppleId = 'AAPL'
MetaID = 'META'
GoogleID = 'GOOGL'
mode = 'TIME_SERIES_DAILY'
```

```python
IBM = AlphaVantage.Stock(IBMId,mode,Key)
IBM_data=IBM.df
IBM_data.time = pd.to_datetime(IBM_data.time)
IBM_data.open = pd.to_numeric(IBM_data.open)
IBM_data.high = pd.to_numeric(IBM_data.high)
IBM_data.close = pd.to_numeric(IBM_data.close)
IBM_data.volume = pd.to_numeric(IBM_data.volume)
IBM_data
```

```python
Meta = AlphaVantage.Stock(MetaID,mode,Key)
Meta_data=Meta.df
Meta_data.time = pd.to_datetime(Meta_data.time)
Meta_data.open = pd.to_numeric(Meta_data.open)
Meta_data.high = pd.to_numeric(Meta_data.high)
Meta_data.close = pd.to_numeric(Meta_data.close)
Meta_data.volume = pd.to_numeric(Meta_data.volume)
Meta_data
```

```python
Google = AlphaVantage.Stock(GoogleID,mode,Key)
Google_data=Google.df
Google_data.time = pd.to_datetime(Google_data.time)
Google_data.open = pd.to_numeric(Google_data.open)
Google_data.high = pd.to_numeric(Google_data.high)
Google_data.close = pd.to_numeric(Google_data.close)
Google_data.volume = pd.to_numeric(Google_data.volume)
Google_data
```

Grouping prices by month to find any seasonal pattern

```python
IBM_data_m = IBM_data
IBM_data_m['month'] = IBM_data_m['time'].dt.month
IBM_avarage_month_price = IBM_data_m.groupby('month', as_index=False)['close'].mean()
sns.barplot(x=IBM_avarage_month_price['month'], y=IBM_avarage_month_price['close'], data=IBM_avarage_month_price)
```

```python
Apple_data_m = Apple_data
Apple_data_m['month'] = Apple_data_m['time'].dt.month
Apple_avarage_month_price = Apple_data_m.groupby('month', as_index=False)['close'].mean()
sns.barplot(x=Apple_avarage_month_price['month'], y=Apple_avarage_month_price['close'], data=Apple_avarage_month_price)
```

```python
Meta_data_m = Meta_data
Meta_data_m['month'] = Meta_data_m['time'].dt.month
Meta_avarage_month_price = Meta_data_m.groupby('month', as_index=False)['close'].mean()
sns.barplot(x=Meta_avarage_month_price['month'], y=Meta_avarage_month_price['close'], data=Meta_avarage_month_price)
```

```python
Google_data_m = Google_data
Google_data_m['month'] = Google_data_m['time'].dt.month
Google_avarage_month_price = Google_data_m.groupby('month', as_index=False)['close'].mean()
sns.barplot(x=Google_avarage_month_price['month'], y=Google_avarage_month_price['close'], data=Google_avarage_month_price)
```

Calculating and plotting moving averages

```python
IBM_moving_averages = IBM_data
IBM_moving_averages['50-day SMA'] = IBM_moving_averages['close'].rolling(window=50).mean()
IBM_moving_averages['200-day SMA'] = IBM_moving_averages['close'].rolling(window=200).mean()

# Plot the closing prices and the moving average
plt.figure(figsize=(10, 6))
plt.plot(IBM_moving_averages['time'], IBM_moving_averages['close'], label='Closing Price')
plt.plot(IBM_moving_averages['time'], IBM_moving_averages['50-day SMA'], label='50-day SMA', linestyle='dotted')
plt.plot(IBM_moving_averages['time'], IBM_moving_averages['200-day SMA'], label='200-day SMA', linestyle='dashed')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price and 200-day Moving Average')
plt.legend()
plt.show()
```

IBM isn`t the best company to demonstrate some trends with SMA methods
